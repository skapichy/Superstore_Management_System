# UNIVARIATE ANALYSIS

### Univariate Analysis is the simplest form of statistiscal Analysis, it is analysis for single variable
### Unit means 'one' and 'variate' mean Variable 

```python
# We want to check how the data are distributed across the categorical columns

categorical_cols = ['customer_segment', 'category', 'region', 'payment_mode', 'delivery', 'auto_reorder']

for col in categorical_cols:
    print(f"\n ... {col}...")
    print(data[col].value_counts())
```

```python
# We want to visualize the distribution of the categorical columns

categorical_cols = ['customer_segment', 'category', 'region', 'payment_mode', 'delivery', 'auto_reorder']

for col in categorical_cols:

    plt.figure(figsize=(10,8))
    data[col].value_counts().plot(kind='bar', color='maroon')
    plt.title(f"Data Distribution of {col}")
    plt.xticks(rotation=30)
    plt.tight_layout()
    plt.show()
```

## Numerical Analysis Variable

```python
# We want to visualize the distribution of the Numerical columns
# Also checking the descriptives statistics of the data e.g mean, median, etc

numerical_cols = ['quantity', 'unit_price', 'discount', 'sales_amount', 'cost_price', 'stock_left', 'reorder_quantity']

for col in numerical_cols:
    print(f"\n.... {col} ....\n")
    print(f"Mean: {data[col].mean():.2f}")
    print(f"Median: {data[col].median():.2f}")
    print(f"Mean: {data[col].mean():.2f}")
    print(f"Std Dev: {data[col].std():.2f}")
    print(f"Min: {data[col].min():.2f}")
    print(f"Max: {data[col].max():.2f}")

    plt.figure(figsize=(8, 4))
    plt.subplot(121)
    data[col].hist(bins=30, color='lightgreen', alpha =0.7)
    plt.title(f"Distribution of {col}")

    plt.subplot(122)
    data.boxplot(column=col)
    plt.title(f"Boxplot of {col}")

    plt.tight_layout()
    plt.show()
```

# BIVARIATE ANALYSIS

### Category Performance of the dataset

```python
# Bivariate Analysis is the statistical analysis of the two variables at the same time

print("\n *** SALES & PROFIT BY CATEGORY ***\n")

category_performance = (
    data.groupby("category")
        .agg({
            "sales_amount": "sum",
            "profit": "sum",
            "order_id": "count"
        })
        .rename(columns={"order_id": "order_count"}))
category_performance['profit_margin'] = (category_performance['profit'] / category_performance['sales_amount']) * 100
print(category_performance, "\n\n")

# Visualization

fig, axes = plt.subplots(1, 2, figsize=(15, 12))

# Sales by Category
category_performance['sales_amount'].plot(
    kind='bar',
    ax=axes[0],
    color='blue',
    alpha=0.7
)
axes[0].set_title("Sales by Category")

# Profit by Category
category_performance['profit'].plot(
    kind='bar',
    ax=axes[1],
    color='green',
    alpha=0.7
)
axes[1].set_title("Profit by Category")

```

```python
# Profit Margin

fig, axes = plt.subplots(1, 2, figsize=(15, 12))
category_performance['profit_margin'].plot(
    kind='bar',
    ax=axes[0],
    color='maroon',
    alpha=0.7
)
axes[0].set_title("Profit Margin")

# Order Count
category_performance['order_count'].plot(
    kind='line',
    ax=axes[1],
    color='maroon',
    alpha=0.7)

plt.tight_layout()
plt.show()
```

### Regional performance

```python
print("\n *** REGIONAL PERFORMANCE***\n")

regional_performance = data.groupby("region").agg({
        "sales_amount": "sum",
        "profit": "sum",
        "order_id": "count"
    }).rename(columns = {"order_id": "order_count"})
regional_performance["profit_margin"] = (regional_performance["profit"] / regional_performance["sales_amount"]) * 100
print(regional_performance)

# Visualization
plt.figure(figsize=(10, 8))

plt.subplot(221)
regional_performance['sales_amount'].plot(kind ='pie', autopct='%1.1f%%')
plt.title('Sales Distribution by Region')

plt.subplot(222)
sns.barplot(x=regional_performance.index, y='order_count', data=regional_performance)
plt.title('Order Count by Region')
plt.xticks(rotation=30)

plt.subplot(223)
sns.barplot(x=regional_performance.index, y='order_count', data=regional_performance)
plt.title('Profit Margin by Region')
plt.xticks(rotation=30)

plt.tight_layout
plt.show()
```

### TIME SERIES ANALYSIS

```python
print("\n *** TIME SERIES ANALYSIS***\n")

# Extract the components
data['order_month'] = data['order_date'].dt.to_period('M')
data['order_year'] = data['order_date'].dt.year

# Monthly Trends
monthly_trends = data.groupby("order_month").agg({
        "sales_amount": "sum",
        "profit": "sum",
        "order_id": "count"
    }).rename(columns = {"order_id": "order_count"})

print('Monthly Trends:')
print(monthly_trends.tail(10), '\n\n')
print(monthly_trends.head(10), '\n\n')

```

```python
# Visualization

plt.figure( figsize=(10,8))

# Monthly Trend by Sales Amount
plt.subplot(311)
monthly_trends['sales_amount'].plot(kind='line', color='black', marker='*')
plt.title('Monthly Sales Trend')
plt.ylabel('Sales Amount')
plt.grid(True)

# Monthly Trend by Profit
plt.subplot(312)
monthly_trends['profit'].plot(kind='line', color='maroon', marker='o')
plt.title('Monthly Profit Trend')
plt.ylabel('Profit')
plt.grid(True)

# Monthly Trend by Order Count
plt.subplot(313)
monthly_trends['order_count'].plot(kind='line', color='blue', marker='+')
plt.title('Monthly Order Count Trend')
plt.ylabel('Order Count')
plt.grid(True)

plt.tight_layout()
plt.show()
```

# ADVANCED ANALYSIS

```python
print("\n *** CORRELATION ANALYSIS ***\n")

# select numerical columns for correlation
numerical_data = data[['quantity', 'unit_price', 'discount', 'sales_amount', 'cost_price', 'profit']]

# Correlation matrix
correlation_matrix = numerical_data.corr()
print("Correlation Matrix:")
print(correlation_matrix, '\n\n')

# Heatmap Visual
plt.figure(figsize=(10,8))
sns.heatmap(correlation_matrix, annot=True, cmap='viridis', center=0, square=True, linewidths=0.5)
plt.title('Correlation Heatmap')
plt.tight_layout()
plt.show()
            
```

### Customer Segment Analysis

```python
# We want to group the customers into segment in doing so creating columns
print("\n *** CUSTOMER SEGMENT ANALYSIS ***\n")

segment_analysis = data.groupby('customer_segment').agg({
        "sales_amount": ["sum",'mean'],
        "profit": ["sum", 'mean'],
        "order_id": "count",
        "discount": "mean"}).round(2)
segment_analysis.columns = ['total_sales', 'avg_sales', 'total_profit', 'avg_profit', 'order_count', 'avg_discount']
print(segment_analysis, '\n\n')
```

```python
# Visualization using the above code

fig, axes =plt.subplots(2,2, figsize=(15, 10))


# Total Sales by Customer segment
segment_analysis['total_sales'].plot(kind='bar', ax=axes[0,0], color='lightblue')
axes[0,0].set_title("Total Sales by Customer Segment")
axes[0,0].tick_params(axis='x', rotation=30)

# Total Profit by Customer segment
segment_analysis['total_profit'].plot(kind='bar', ax=axes[0,1], color='maroon')
axes[0,1].set_title("Total Profit by Customer Segment")
axes[0,1].tick_params(axis='x', rotation=30)

# Average Profit by Customer segment
segment_analysis['avg_profit'].plot(kind='bar', ax=axes[1,0], color='green')
axes[1,0].set_title("Average Profit by Customer Segment")
axes[1,0].tick_params(axis='x', rotation=30)

# Average Discount by Customer segment
segment_analysis['avg_discount'].plot(kind='bar', ax=axes[1,1], color='chocolate')
axes[1,1].set_title("Average Discount by Customer Segment")
axes[1,1].tick_params(axis='x', rotation=30)

plt.tight_layout
plt.show()
```

# KEY INSIGHTS & SUMMARY

```python
# Overall Metrics

total_sales = data['sales_amount'].sum()
total_profit = data['profit'].sum()
total_orders = data['order_id'].nunique()
avg_profit_margin = (total_profit/total_sales) * 100

print(f'OVERALL PERFORMANCE:\n')
print(f'Total Sales: ${total_sales:,.2f}')
print(f'Total Profit: ${total_profit:,.2f}')
print(f'Total Orders: {total_orders:}')
print(f'Average Profit Margin: {avg_profit_margin:,.2f}%')
```

```python
# Top Performing Categories
top_categories = data.groupby('category')['profit'].sum().sort_values(ascending=False)
print(f'\n TOP PERFORMING CATEGORIES: \n')
for i, (category, profit) in enumerate(top_categories.items(),1):
    print(f'{i}.{category}: ${profit:,.2f}')


# Regional Performance
best_region = data.groupby('region')['profit'].sum().idxmax()
print(f'\n BEST PERFORMING REGION: {best_region} \n')

# Delivery Performance
delivery_stats = data['delivery'].value_counts(normalize=True) * 100
print(f'\n DELIVERY STATUS \n')
for status, percentage in delivery_stats.items():
    print(f'{status}: {percentage: .1f}%')

# Customer segments
best_segment = data.groupby('customer_segment')['profit'].sum().idxmax()
print(f'\n BEST CUSTOMER SEGMENT: {best_segment}')
    
```

# EXPORT RESULTS

```python
analysis_results = {
    'total_sales': total_sales,
    'total_profit': total_profit,
    'total_orders': total_orders,
    'avg_profit_margin': avg_profit_margin,
    'top_category': top_categories.index[0],
    'best_region' : best_region,
    'best_customer_segment': best_segment}

# Convert to DataFrame and export to save
result_data = pd.DataFrame([analysis_results])
result_data.to_csv('Exploratory_data_analysis_summary.csv', index=False)
print('Analysis save and good to use.')
```
